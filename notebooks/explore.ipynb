```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
DATASETS = {
    'PlantVillage': r'D:\crop-disease-advisor\data\plantvillage dataset\color',
    'Sugarcane':    r'D:\crop-disease-advisor\data\sugarcane',
    'Groundnut':    r'D:\crop-disease-advisor\data\groundnut',
    'Rice':         r'D:\crop-disease-advisor\data\rice',
    'Wheat':        r'D:\crop-disease-advisor\data\wheat',
}
rows = []
for dataset_name, path in DATASETS.items():
    for cls in os.listdir(path):
        cls_path = os.path.join(path, cls)
        if os.path.isdir(cls_path):
            count = len(os.listdir(cls_path))
            rows.append({
                'dataset': dataset_name,
                'class_name': cls,
                'image_count': count
            })
df = pd.DataFrame(rows)
print(df.shape)
print(df.head())
```

```python
plt.figure(figsize=(24, 8))


df_sorted = df.sort_values(by='image_count', ascending=True)


palette = sns.color_palette("Set2", n_colors=df['dataset'].nunique())


sns.barplot(
    data=df_sorted,
    x='class_name',
    y='image_count',
    hue='dataset',
    palette=palette
)

plt.xticks(rotation=90, fontsize=6)
plt.title("Class Distribution Across All Datasets (73 Classes)")
plt.ylabel("Number of Images")
plt.tight_layout()

plt.savefig(
    r"D:\crop-disease-advisor\notebooks\class_distribution.png",
    dpi=150
)

plt.show()


print("10 smallest classes:")
print(df_sorted.head(10))
```
