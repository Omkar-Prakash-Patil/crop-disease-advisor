```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
DATASETS = {
    'PlantVillage': r'D:\crop-disease-advisor\data\plantvillage dataset\color',
    'Sugarcane':    r'D:\crop-disease-advisor\data\sugarcane',
    'Groundnut':    r'D:\crop-disease-advisor\data\groundnut',
    'Rice':         r'D:\crop-disease-advisor\data\rice',
    'Wheat':        r'D:\crop-disease-advisor\data\wheat',
}
rows = []
for dataset_name, path in DATASETS.items():
    for cls in os.listdir(path):
        cls_path = os.path.join(path, cls)
        if os.path.isdir(cls_path):
            count = len(os.listdir(cls_path))
            rows.append({
                'dataset': dataset_name,
                'class_name': cls,
                'image_count': count
            })
df = pd.DataFrame(rows)
print(df.shape)
print(df.head())
```
